```python
from pandas.io.parsers import *
import numpy as np
import mysql.connector
import pandas as pd
from pandas.io.sql import *
cnx = mysql.connector.connect(user='davidc', password='mysql_password', host='127.0.0.1',  database='CRISPR')
```

```python
cursor = cnx.cursor()
letter_to_idx = {'A':0,'C':1,'G':2,'T':3}
for runID in [26, 27, 28, 29]:
    counts = np.zeros((8,4))
    cursor.execute('select barcode from samples where runID=%s',(runID,))
    for (barcode,) in cursor:
        for j in range(len(barcode)):
            counts[j, letter_to_idx[barcode[j]]] += 1
    df = pd.DataFrame(counts, index=range(1,9))
    df.columns = ['A','C','G','T']
    df = df.apply(lambda x:x/sum(x), 1)
    df.plot(kind='bar',stacked=True,legend=True)
    legend(bbox_to_anchor=(1.05, 1),loc=2)
```

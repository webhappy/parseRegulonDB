```python
def get_all_timepoints (condition, replicate):  # condition is e.g. ab_aero
    return [all_samples[condition+'_t30_'+str(replicate)], all_samples[condition+'_t60_'+str(replicate)],
            all_samples[condition+'_t180_'+str(replicate)]]
```

#### Prepare summed_counts as dataframe where counts for all time points are added together

```python
# Grab all counts
summed_counts = None  # This becomes a dataframe with 18 columnsa
for cond in all_conditions:
    for rep in ['1','2']:
        t = read_sql('select sgRNA_seq, SUM(count) AS %s from counts where sampleID in (%s) GROUP BY sgRNA_seq '%
                     (cond+'_'+rep,','.join(map(str,get_all_timepoints(cond, rep)))),cnx,'sgRNA_seq')
        if summed_counts is None:
            summed_counts = t
        else:
            summed_counts = summed_counts.join(t)
print "Finished making summed_counts with shape ",summed_counts.shape
```

```python
# Make proportions for a sample_ID
def calc_proportions (sampleID):
    t = read_sql('select count, sgRNA_seq from counts where sampleID=%s'%sampleID,cnx,'sgRNA_seq')
    assert t.shape == (NUM_SGRNA,1)
    t.sort_index(inplace=True)
    t = t + 1  # convert to pseudo-count
    t = t.apply(lambda x:x/sum(x),0) # normalize by sum of this column
    return t
```

# Prepare scores per gene (choose one method within here)
The goal is to prepare the df that is a DataFrame that contains one column per condition and one row per gene

```python
sgRNAs_with_gene = read_frame('select seq, gene_name from sgRNAs where gene_name!=\'\'', cnx, index_col='seq')
```

### Prepare scores by using t180 proportions and taking mean of the two replicates

```python
# Calculate scores for all sgRNA's over all conditions
scores = np.zeros((NUM_SGRNA, len(all_treatments)))
assert len(treatments)*2 == len(all_treatments)
i = 0
for cond in treatments:  # take the mean of the two replicates
    for ae in ['aero', 'anaero']:
        #print cond,ae,'uses',cond+'_'+ae+'_'+'t180_1','->',all_samples[cond+'_'+ae+'_'+'t180_1'],'and','cont'+'_'+ae+'_'+'t180_1','->',all_samples['cont'+'_'+ae+'_'+'t180_1']
        rep1_proportion = np.log2( calc_proportions(all_samples[cond+'_'+ae+'_'+'t180_1']) 
                                  / calc_proportions(all_samples['cont'+'_'+ae+'_'+'t180_1']))
        rep2_proportion = np.log2( calc_proportions(all_samples[cond+'_'+ae+'_'+'t180_2'])
                                  / calc_proportions(all_samples['cont'+'_'+ae+'_'+'t180_2']))
        score = ((rep1_proportion + rep2_proportion)/2 )
        scores[:,i] = score.iloc[:,0]
        i += 1 # Increment the column we're using
        
# Now get median score per gene
df = pd.DataFrame(scores,score.index, all_treatments)
df = df.join(sgRNAs_with_gene)
df = df.groupby('gene_name').median()
```

### Prepare scores by summing up all counts for all sgRNA's for each gene

```python
summed_counts_with_gene = summed_counts.join(sgRNAs_with_gene).groupby('gene_name').sum()
summed_counts_with_gene += 1  # In case we have 0 counts for a gene, make it a psuedo-count so we don't attempt a division by 0
scores = np.zeros((summed_counts_with_gene.shape[0], len(all_treatments)))
assert len(treatments)*2 == len(all_treatments)
i = 0

for cond in treatments:  # take the mean of the two replicates
    for ae in ['aero', 'anaero']:
        rep1_score = np.log2(summed_counts_with_gene[cond+'_'+ae+'_1'] / summed_counts_with_gene['cont'+'_'+ae+'_1'])
        rep2_score = np.log2(summed_counts_with_gene[cond+'_'+ae+'_2'] / summed_counts_with_gene['cont'+'_'+ae+'_2'])
        scores[:,i] = (rep1_score + rep2_score)/2
        i += 1 # Increment the column we're using
df = pd.DataFrame(scores, summed_counts_with_gene.index, all_treatments)
```

## Perform the clustering

### Perform hierarchical clustering and save this to a file in case we want to look at these results in JTV

```python
import Bio.Cluster as Pycluster
#hierarchical = treecluster(df.values, method='a')
record = Pycluster.read(open('out.txt'))
tree = record.treecluster(method='a')
tree.scale()
record.save('temp_hierarchical_genes',tree)
```

```python
# using tree to initialize clustering
def generate_clusters (data, num_clusters, return_wcss = False):
    '''Returns a list with num_clusters items where each item is a list with a row index
    data is a matrix with rows for genes/sgRNA's and columsn representing conditions. 
    num_clusters is an int.
    return_wcss is a boolean to indicate if we should return a tuple that also indicates the WCSS score
    '''
    cluster_ids, wcss, ignore = Pycluster.kcluster(data, nclusters=num_clusters, initialid=tree.cut(num_clusters))
    # Now convert to list of lists
    clusters = []
    for j in xrange(num_clusters):
        clusters.append(list())

    for k in xrange(len(cluster_ids)):
            clusters[cluster_ids[k]].append(k)
    if return_wcss:
        return clusters, wcss
    else:
        return clusters
```

### Generate clusters and store into the results/n<#>/ with each txt file holding a list of genes

```python
import os, shutil
wcss_scores = []
for num_clusters in range(2,21):
    clusters, wcss = generate_clusters(df.values, num_clusters, True)
    wcss_scores.append(wcss)
    cur_dir = 'results/n'+str(num_clusters)+'/'
    if os.path.isdir(cur_dir):
        shutil.rmtree(cur_dir)
    os.makedirs(cur_dir)
    for k in xrange(len(clusters)):
        cur_file = cur_dir + 'cluster'+str(k)+'.txt'
        cur_file = open(cur_file,'w')
        for j in xrange(len(clusters[k])):
            cur_gene = df.index[clusters[k][j]]
            cur_file.write(cur_gene + '\n')
        cur_file.close()
```

```python
# FOM
def calc_FOM_for_condition (num_clusters, condition, scores):
    '''Calculates reduced data matrix to send into K-means to generate FOM
    num_clusters is int
    condition is int
    scores is 2-D matrix with all data
    '''
    temp = 0.0
    # generate clusters by deleting the data
    reduced = np.delete(scores, condition, 1)
    clusters = generate_clusters(reduced, num_clusters)
    for c in xrange(len(clusters)):
        ave_score = np.average(scores[c,:])
        for g in xrange(len(clusters[c])):
            temp += (scores[g, c] - ave_score)**2
    temp /= scores.shape[1]
    return np.sqrt(temp * (scores.shape[1]-num_clusters)/scores.shape[1])

def calc_FOM (num_clusters, scores):
    sum_of_FOM = 0.0
    for cond in xrange(scores.shape[1]):
        sum_of_FOM += calc_FOM_for_condition(num_clusters, cond, scores)
    return sum_of_FOM

FOM_scores = [calc_FOM(x, df.values) for x in range(1,11)]
```

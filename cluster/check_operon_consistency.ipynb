```python
import pandas.io.sql as psql
import numpy as np
import pandas as pd
import csv, mysql.connector

con=mysql.connector.connect(user='davidc', password='mysql_password', host='127.0.0.1', database='CRISPR')
column_names_rep1 = {'control':('1_aerobic_t30_1_control', '1_aerobic_t60_1_control', '1_aerobic_t180_1_control'),
              'chlor':('1_aerobic_t30_2_chlor', '1_aerobic_t60_2_chlor', '1_aerobic_t180_2_chlor'),
              'nor':('1_aerobic_t30_6_nor', '1_aerobic_t60_6_nor', '1_aerobic_t180_6_Nor'),
              }
column_names_rep2 = {'control':('2_aerobic_Control_t30_2','2_aerobic_Control_t60_2','2_aerobic_Control_t180_2'),
              'chlor':('2_aerobic_Chlor_t30_2','2_aerobic_Chlor_t60_2','2_aerobic_t180_2_chlor'),
              'nor':('2_aerobic_Nor_t30_2','2_aerobic_Nor_t60_2','2_aerobic_t180_6_nor')}
conditions = ['chlor','nor']  # assume first entry at index 0 is the control
```

```python
# Assign scores per gene
df = pd.io.parsers.read_csv(open('selected.csv'))
df.index = df['seq']
NUM_SGRNA = len(df.iloc[:,0])
proportions = (df.iloc[:, 1:] + 1).iloc[:,:].apply(lambda x:x/sum(x),1)  # Normalize to proportions
sgRNAs_with_gene = psql.read_frame('select * from sgRNAs where gene_name!=" "', con, index_col='seq')
df = df.merge(sgRNAs_with_gene, 'left', left_index=True, right_index=True)
```

```python
genes = df.gene_name.dropna().unique()
scores1 = {}
scores2 = {}
for gene in genes:
    scores1[gene]=[]
    scores2[gene]=[]
    rows = df[df.gene_name==gene]
    for c in conditions:
        for k in xrange(3):  # Use median of sgRNA's for each gene
            scores1[gene].append(np.log2(rows[column_names_rep1[c][k]] / rows[column_names_rep1['control'][k]]).median())
            scores2[gene].append(np.log2(rows[column_names_rep2[c][k]] / rows[column_names_rep2['control'][k]]).median())
scoresDF1 = pd.DataFrame.from_dict(scores1, orient='index')
scoresDF1.columns = [column_names[c][k] for c in conditions for k in xrange(3)]
scoresDF2 = pd.DataFrame.from_dict(scores2, orient='index')
scoresDF2.columns = [column_names[c][k] for c in conditions for k in xrange(3)]
```
